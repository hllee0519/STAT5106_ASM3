```python
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=[random.SystemRandom().uniform(0, 1)for _ in range(100000)]
o=[random.SystemRandom().uniform(0, 2*np.pi)for _ in range(100000)]
```

```python
x_end=[]
for i in range(len(x)):
  x_end.append(((x[i]+(np.cos([o[i]])[0]/2), x[i]-(np.cos([o[i]])[0]/2))))
```

```python
max_larger_than_1=[]
min_smaller_than_0=[]
for i in range(len(x_end)):
    max_larger_than_1.append(max(x_end[i])>0)
    min_smaller_than_0.append(min(x_end[i])<0)
result=np.logical_and(max_larger_than_1, min_smaller_than_0)
new_pi=1/(result.tolist().count(True)/len(result))
new_pi
```

```python
new_pi_list=[]
for i in range(100):
    x=[random.SystemRandom().uniform(0, 1)for _ in range(100000)]
    o=[random.SystemRandom().uniform(0, 2*np.pi)for _ in range(100000)]
    x_end=[]
    for i in range(len(x)):
        x_end.append(((x[i]+(np.cos([o[i]])[0]/2), x[i]-(np.cos([o[i]])[0]/2))))
    max_larger_than_1=[]
    min_smaller_than_0=[]
    for i in range(len(x_end)):
        max_larger_than_1.append(max(x_end[i])>0)
        min_smaller_than_0.append(min(x_end[i])<0)
    result=np.logical_and(max_larger_than_1, min_smaller_than_0)
    new_pi_list.append(1/(result.tolist().count(True)/len(result)))
```

```python
df=pd.DataFrame(data={"pi":new_pi_list})
df['mean']=df.pi.mean()
df.plot(y=["pi", 'mean'],figsize=(20,8))
```
